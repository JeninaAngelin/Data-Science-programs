```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the Titanic dataset from seaborn
titanic = sns.load_dataset("titanic")
```

```python
# Create a figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
plt.subplots_adjust(hspace=0.4)
# Binned Age Distribution
axes[0, 0].hist(titanic['age'].dropna(), bins=20, edgecolor='black')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].set_title('Binned Age Distribution')
# Binned Fare Distribution
axes[0, 1].hist(titanic['fare'], bins=20, edgecolor='black')
axes[0, 1].set_xlabel('Fare')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].set_title('Binned Fare Distribution')
# Binned Age vs. Fare with Hexbin
sns.scatterplot(x='age', y='fare', data=titanic, ax=axes[1, 0], alpha=0.6)    #hexbin
axes[1, 0].set_xlabel('Age')
axes[1, 0].set_ylabel('Fare')
axes[1, 0].set_title('Binned Age vs. Fare with Hexbin')
# Categorical Binning - Bar Plot for Class
sns.countplot(x='class', data=titanic, ax=axes[1, 1], palette='Set3')        #barplot
axes[1, 1].set_xlabel('Class')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].set_title('Categorical Binning - Class Bar Plot')
plt.tight_layout()
plt.show()
```
