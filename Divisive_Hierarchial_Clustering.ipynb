### Ward's minimum variance method

In statistics, Ward's method is a criterion applied in hierarchical cluster analysis. With hierarchical clustering, the sum of squares starts out at zero (because every point is in its own cluster) and then grows as we merge clusters. Ward's method keeps this growth as small as possible. The Euclidean distance is the “ordinary” straight-line distance between two points in Euclidean space.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Create a small synthetic dataset
X = np.array([[1, 2], [2, 2], [2, 3], [8, 8], [9, 8], [9, 9]])
# Compute linkage matrix for divisive hierarchical clustering
Z = linkage(X, method='ward')
```

```python
# Plot dendrogram
plt.figure()
dendrogram(Z)
plt.title("Divisive Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```
