Agglomerative:

Initially, each object is considered to be its own cluster. According to a particular procedure, the clusters are then merged step by step until a single cluster remains.

Divisive:

The Divisive method is the opposite of the Agglomerative method.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Generate a synthetic dataset with clusters
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)
```

```python
# Perform clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X)
```

```python
# Create histograms for each cluster
plt.figure(figsize=(10, 6))
for cluster in range(4):
 cluster_data = X[kmeans_labels == cluster]
 plt.hist(cluster_data[:, 0], alpha=0.5, label=f'Cluster {cluster + 1}, Feature 1')
 plt.hist(cluster_data[:, 1], alpha=0.5, label=f'Cluster {cluster + 1}, Feature 2')
plt.xlabel('Feature Value')
plt.ylabel('Frequency')
plt.title('Histograms of Clusters')
plt.legend()
plt.show()
```
