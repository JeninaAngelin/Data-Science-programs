the Haar wavelet is a sequence of rescaled "square-shaped" functions which together form a wavelet family or basis.

```python
def haar_wavelet_transform(data):
 result = []
 while len(data) >= 2:
  average = (data[0] + data[1]) / 2
  difference = (data[0] - data[1]) / 2
  result.append(average)
  result.append(difference)
  data = data[2:]
 return result + data
```

```python
def inverse_haar_wavelet_transform(transformed_data):
 data_length = len(transformed_data)
 original_data = [0] * data_length
 for i in range(0, data_length, 2):
  average = transformed_data[i]
  difference = transformed_data[i + 1]
  original_data[min(2 * i, data_length - 1)] = average + difference
  original_data[min(2 * i + 1, data_length - 1)] = average - difference
 return original_data
```

```python
data = [5, 10, 3, 8, -2, 6, 1, 4]
transformed_data = haar_wavelet_transform(data)
print("Transformed Data:", transformed_data)
original_data = inverse_haar_wavelet_transform(transformed_data)
print("Original Data:", original_data)
```
